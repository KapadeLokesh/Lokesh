```python
with open("sample.json", "r") as file:
    data = json.load(file)

data
```

```python
# 2) Preprocess the data:
#   * Remove duplicate entries using Python's set.
#   * Use regular expressions to clean up phone numbers by enforcing a consistent format: (XXX) XXX-XXXX.eg- (123) 456-7890
#   * Standardize the email format to lowercase.
#   * Ensure each name starts with a capital letter.

for i in data:
    i['email'] = i['email'].lower()

data
```

```python
for i in data:
    i['name'] = i['name'].title()

data
```

```python
import re
 
def clean_phoneNo(phone):
    digits = re.sub(r'\D', '', phone)
    if len(digits) == 11:
        digits = digits[1:]
    if len(digits) == 12:
        digits = digits[2:]
    if len(digits) == 10:
        return f"({digits[:3]}) {digits[3:6]}-{digits[6:]}"

for i in data:
    i['phone_number'] = clean_phoneNo(i['phone_number'])
    
data
```

```python
# 3) Add complexity by including nested data:
#   * Each address should be split into street, city, state, and zip_code.
#   * Extract the state abbreviation from the address field.

import re
 
def split_address(address):
    match = re.match(r"^(.*),\s*(.*),\s*([A-Z]{2})\s*(\d{5})$", address)
    if match:
        street, city, state, zip_code = match.groups()
        return {
            "street": street.strip(),
            "city": city.strip(),
            "state": state.strip(),
            "zip_code": zip_code.strip(),
        }
    else:
        return {"street": "", "city": "", "state": "", "zip_code": ""}  # Handle invalid addresses
 
 
for record in data:
    record["address"] = split_address(record["address"]) 
```
